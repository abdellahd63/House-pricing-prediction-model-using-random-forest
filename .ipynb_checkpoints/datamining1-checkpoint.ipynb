#### CREATING A SET OF UNSEEN DATA FOR FINAL TESTING

```python
# Shuffling
dfnonNULL = dfnonNULL.sample(frac=1).reset_index(drop=True)
```

```python
# Calculate the split index
split_idx = int(0.85 * len(dfnonNULL))

# Split the DataFrame
train_df = dfnonNULL[:split_idx]
test_df = dfnonNULL[split_idx:]
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
# Will be used lastly to test the model
final_x_test=test_df.drop(['median_house_value'] , axis=1)
```

```python
from sklearn.model_selection import train_test_split
x = train_df.drop(['median_house_value'] , axis=1)
y = train_df['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2)
```

```python
train_data=x_train.join(y_train) 
```

```python
train_data
```

```python
from sklearn.linear_model import LinearRegression

x_train , y_train = train_data.drop(["median_house_value"], axis = 1 ), train_data["median_house_value"]

reg = LinearRegression()

reg.fit(x_train , y_train)
```

```python
test_data=x_test.join(y_test) 

test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data["total_bedrooms"]=np.log(test_data["total_bedrooms"]+1)
test_data["population"]=np.log(test_data["population"]+1)
test_data["households"]=np.log(test_data["households"]+1)
test_data["median_income"]=np.log(test_data["median_income"]+1)

test_data['bedrooms_ratio']=test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms'] / test_data['households']
```

```python
reg.score(x_test,y_test)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train , y_train)
```

```python
forest.score(x_test , y_test)
```

```python
from sklearn.model_selection import GridSearchCV
#cross validation with all combin

params_grid={
    "n_estimators":[100,200,300],
    "min_samples_split":[None,4,8],
    "max_depth":[None,4,8]
}

grid_search=GridSearchCV(forest, params_grid,cv=5,scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(x_train , y_train)
```

```python
best_forest=grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```

#### TESTING ON UNSEEN DATA

```python
predictions=best_forest.predict(final_x_test)
```

```python
predictions
```

```python
actuals = test_df['median_house_value']
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate Mean Absolute Error
mae = mean_absolute_error(actuals, predictions)
print("Mean Absolute Error (MAE):", mae)

# Calculate Mean Squared Error
mse = mean_squared_error(actuals, predictions)
print("Mean Squared Error (MSE):", mse)

# Calculate Root Mean Squared Error
rmse = mean_squared_error(actuals, predictions, squared=False)
print("Root Mean Squared Error (RMSE):", rmse)

# Calculate R-squared
r2 = r2_score(actuals, predictions)
print("R-squared (R²):", r2)
```
